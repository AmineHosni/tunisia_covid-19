```python
df = pd.read_excel("2020-04-11_JHU.xlsx")
```

```python
tun = df[df.Country=='Tunisia']
```

```python
tun.head()
```

```python
tun_indexed = tun.set_index('Date')
```

```python
tun_indexed.head()
```

```python
tun_indexed['Confirmed_evolution'] = 100*(tun_indexed.Confirmed-tun_indexed.Confirmed.shift())/tun_indexed.Confirmed.shift() # (2 - 1)/2
```

```python
tun_indexed.loc[tun_indexed[tun_indexed.Confirmed>0].index[0]:].head(1)
```

```python

tun_indexed.loc[tun_indexed[tun_indexed.Confirmed>0].index[0]:].tail(1)
```

```python
tun_indexed['new_cases'] = (tun_indexed.Confirmed-tun_indexed.Confirmed.shift()) # (2 - 1)/2
```

```python
tun_indexed['Deaths_evolution'] = (tun_indexed.Deaths-tun_indexed.Deaths.shift())/tun_indexed.Deaths.shift() # (2 - 1)/2
```

```python
import numpy as np
tun_log = tun.copy()

for c in tun_log.columns[1:-1]:
    tun_log[c] = tun_log[c].apply(np.log)
```

```python
tun_log.fillna(0,inplace=True)
```

```python
tun_indexed.columns
```

```python
tun = tun_indexed.copy()
tun = tun.reset_index()


# tun['Date'] = tun.index
tun = tun[['Date','Confirmed','new_cases','Confirmed_evolution','Deaths','Deaths_evolution','Recovered']]
tun.fillna(0,inplace = True)
```

```python
import plotly.graph_objects as go


def plot_cases(df, rangeslider_visible=True,my_title=False):
    fig = go.Figure()
    
    for col in df.columns[1:]:
    
        fig.add_trace(go.Scatter(x=df.Date, y=df[col],
                            mode='lines+markers',
                            name=col, connectgaps=True))
    fig.update_xaxes(rangeslider_visible=True)

    if my_title:
        fig.update_layout(title=my_title)
    fig.show()
```

```python
plot_cases(tun)
```

```python
plot_cases(tun_log, my_title="Logarithmic scale")
```
